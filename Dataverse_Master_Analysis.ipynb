# Urban Heat Island - Master Analysis

This notebook contains the consolidated data science workflow for the Dataverse Hackathon. We utilize advanced machine learning and statistical modeling to analyze the Urban Heat Island effect.

## 1. Spatial Clustering (K-Means)
We use `scikit-learn` to cluster neighborhoods into risk zones based on structural characteristics like tree cover, asphalt density, and building density.

## 2. Time Series Forecasting
We forecast the composite Urban Heat Risk Index using two approaches:
- **SARIMAX**: A traditional statistical approach handling seasonality and exogenous weather variables.
- **Random Forest Regressor**: A modern machine learning approach utilizing lag features to capture non-linear temporal dynamics.

## 3. Health Impact Modeling
We analyze the impact of heat on public health using **Poisson Regression (Generalized Linear Models)**. Since health outcomes (e.g., heatstroke deaths, fatigue cases) are count data, Poisson regression provides a mathematically rigorous model compared to standard OLS.

*Note: The full interactive visualization and dynamic execution of these models are implemented in `dashboard.py`.*
